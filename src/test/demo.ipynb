```python
# Example of using httpx for making HTTP requests
# httpx is a modern HTTP client that supports async/await and has a similar API to requests

# Basic GET request
response = httpx.get('https://httpbin.org/get')
print("Basic GET request status code:", response.status_code)
print("Response JSON:", response.json())

# POST request with JSON data
data = {'key': 'value'}
response = httpx.post('https://httpbin.org/post', json=data)
print("\nPOST request status code:", response.status_code)
print("Response JSON:", response.json())

# Request with custom headers
headers = {'User-Agent': 'Custom User Agent'}
response = httpx.get('https://httpbin.org/headers', headers=headers)
print("\nCustom headers request status code:", response.status_code)
print("Response headers:", response.json())

# Using query parameters
params = {'param1': 'value1', 'param2': 'value2'}
response = httpx.get('https://httpbin.org/get', params=params)
print("\nRequest with query params status code:", response.status_code)
print("Response with params:", response.json())

# Error handling example
try:
    response = httpx.get('https://httpbin.org/status/404')
    response.raise_for_status()
except httpx.HTTPStatusError as e:
    print(f"\nHTTP Error occurred: {e}")

# Using a client for multiple requests with same base URL
with httpx.Client(base_url='https://httpbin.org') as client:
    response = client.get('/get')
    print("\nUsing client response status:", response.status_code)
```
